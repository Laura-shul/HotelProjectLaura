# Introduction

This notebook aims to solve the problem of allocating hotel rooms to guests using different strategies. 
We will implement four strategies:
1. **Random Allocation**
2. **Preference-Based Allocation**
3. **Price-Based Allocation**
4. **Availability-Based Allocation**

Each strategy will be analyzed in terms of total revenue, guest allocation efficiency, and hotel utilization.

# Step 1: Import Libraries and Set Up Environment
In this step, we import all necessary libraries, set up logging for tracking our project's progress, and define the base directory for our data files.

```python
import logging
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Ensure plots are displayed in the notebook
%matplotlib inline

# Set up logging to track project progress
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logging.info('Starting Hotel Allocation Project')

# Define base directory for data files
BASE_DIR = 'data'
```

# Step 2: Load Data
In this step, we load data from Excel files containing information about hotels, guests, and their preferences.
The function `load_and_prepare_data` helps to load data and set the index if needed.
We will load three datasets: `hotels`, `guests`, and `preferences`.

```python
import os

def load_and_prepare_data(base_dir, file_name, index_col=None):
    file_path = os.path.join(base_dir, file_name)
    try:
        df = pd.read_excel(file_path)
        if index_col and index_col in df.columns:
            df.set_index(index_col, inplace=True)
        return df
    except FileNotFoundError:
        logging.error(f"File {file_name} not found.")
        return pd.DataFrame()

# Load data
hotels = load_and_prepare_data(BASE_DIR, 'hotels.xlsx', index_col='hotel')
guests = load_and_prepare_data(BASE_DIR, 'guests.xlsx', index_col='guest')
preferences = load_and_prepare_data(BASE_DIR, 'preferences.xlsx')

# Verify the loaded data
display(hotels.head(), guests.head(), preferences.head())
```

# Step 3: Define Allocation Strategies

We define different strategies to allocate hotel rooms to guests. Each strategy has its unique approach to achieve the goal.
Below we start with the "Random Allocation" strategy, which assigns rooms to guests randomly.

```python
# Random Allocation Strategy
def random_allocation(hotel_data, guest_data):
    hotels_copy = hotel_data.copy()
    allocation = []

    guest_ids = guest_data.index.tolist()
    random.shuffle(guest_ids)

    # Allocate guests randomly to available rooms
    for guest_id in guest_ids:
        for hotel_id, hotel in hotels_copy.iterrows():
            if hotel['rooms'] > 0:
                allocation.append({
                    'guest_id': guest_id,
                    'hotel_id': hotel_id,
                    'final_price': float(hotel['price'] * (1 - guest_data.loc[guest_id, 'discount']))
                })
                hotels_copy.at[hotel_id, 'rooms'] -= 1
                break

    return allocation

# Execute Random Allocation Strategy
random_results = random_allocation(hotels.copy(), guests.copy())

# Convert results to DataFrame
random_results_df = pd.DataFrame(random_results)

# Analysis of Random Allocation Results
total_revenue = random_results_df['final_price'].sum()
total_guests_allocated = len(random_results_df)
unique_hotels_used = random_results_df['hotel_id'].nunique()

# Store results for further use
analysis_results = {}
analysis_results['Random Allocation'] = {
    'revenue': total_revenue,
    'guests_allocated': total_guests_allocated,
    'unique_hotels_used': unique_hotels_used
}

# Visualization of Random Allocation
plt.figure(figsize=(8, 6))

# Adding hue parameter as a workaround for the deprecation warning
sns.barplot(x=['Random Allocation'], y=[total_revenue], palette='viridis', hue=['Random Allocation'])
plt.legend([], [], frameon=False)  # Disable the legend if it is not needed
plt.title('Total Revenue for Random Allocation')
plt.ylabel('Total Revenue ($)')
plt.tight_layout()
plt.show()
```

## Preference Allocation Strategy

This strategy aims to maximize guest satisfaction by assigning guests to hotels based on their preferences.
The guests are processed in sequence, and rooms are allocated according to their preferred hotels.

```python
# Preference-Based Allocation Strategy
def preference_allocation(hotel_data, guest_data, preferences_data):
    hotels_copy = hotel_data.copy()
    allocation = []

    preferences_grouped = preferences_data.groupby('guest')

    for guest_id in guest_data.index:
        allocated = False

        if guest_id in preferences_grouped.groups:
            preferred_hotels = preferences_grouped.get_group(guest_id).sort_values('priority')
            for _, preference in preferred_hotels.iterrows():
                hotel_id = preference['hotel']
                if hotel_id in hotels_copy.index and hotels_copy.loc[hotel_id, 'rooms'] > 0:
                    allocation.append({
                        'guest_id': guest_id,
                        'hotel_id': hotel_id,
                        'final_price': float(hotels_copy.loc[hotel_id, 'price'] * (1 - guest_data.loc[guest_id, 'discount']))
                    })
                    hotels_copy.at[hotel_id, 'rooms'] -= 1
                    allocated = True
                    break

        if not allocated:
            for hotel_id, hotel in hotels_copy.iterrows():
                if hotel['rooms'] > 0:
                    allocation.append({
                        'guest_id': guest_id,
                        'hotel_id': hotel_id,
                        'final_price': float(hotel['price'] * (1 - guest_data.loc[guest_id, 'discount']))
                    })
                    hotels_copy.at[hotel_id, 'rooms'] -= 1
                    break

    return allocation

# Execute Preference Allocation Strategy
preference_results = preference_allocation(hotels.copy(), guests.copy(), preferences.copy())

# Convert results to DataFrame
preference_results_df = pd.DataFrame(preference_results)

# Analysis of Preference Allocation Results
total_revenue = preference_results_df['final_price'].sum()
total_guests_allocated = len(preference_results_df)
unique_hotels_used = preference_results_df['hotel_id'].nunique()

# Store results for further use
analysis_results['Preference Allocation'] = {
    'revenue': total_revenue,
    'guests_allocated': total_guests_allocated,
    'unique_hotels_used': unique_hotels_used
}

# Visualization of Preference Allocation
plt.figure(figsize=(8, 6))

# Adding hue parameter to eliminate the deprecation warning
sns.barplot(x=['Preference Allocation'], y=[total_revenue], palette='plasma', hue=['Preference Allocation'])
plt.legend([], [], frameon=False)  # Disable the legend if it is not needed
plt.title('Total Revenue for Preference Allocation')
plt.ylabel('Total Revenue ($)')
plt.tight_layout()
plt.show()
```

## Price-Based Allocation Strategy

This strategy aims to minimize the cost for guests by allocating rooms starting with the cheapest hotel first.
It can achieve higher occupancy rates while keeping costs low for guests.

```python
# Price-Based Allocation Strategy
def price_allocation(hotel_data, guest_data):
    hotels_copy = hotel_data.sort_values('price').copy()  # Sort hotels by price in ascending order
    allocation = []

    for guest_id in guest_data.index:
        for hotel_id, hotel in hotels_copy.iterrows():
            if hotel['rooms'] > 0:
                allocation.append({
                    'guest_id': guest_id,
                    'hotel_id': hotel_id,
                    'final_price': float(hotel['price'] * (1 - guest_data.loc[guest_id, 'discount']))
                })
                hotels_copy.at[hotel_id, 'rooms'] -= 1
                break

    return allocation

# Execute Price Allocation Strategy
price_results = price_allocation(hotels.copy(), guests.copy())

# Convert results to DataFrame
price_results_df = pd.DataFrame(price_results)

# Analysis of Price Allocation Results
total_revenue = price_results_df['final_price'].sum()
total_guests_allocated = len(price_results_df)
unique_hotels_used = price_results_df['hotel_id'].nunique()

# Store results for further use
analysis_results['Price Allocation'] = {
    'revenue': total_revenue,
    'guests_allocated': total_guests_allocated,
    'unique_hotels_used': unique_hotels_used
}

# Visualization of Price Allocation
plt.figure(figsize=(8, 6))

# Adding hue parameter to eliminate the deprecation warning
sns.barplot(x=['Price Allocation'], y=[total_revenue], palette='magma', hue=['Price Allocation'])
plt.legend([], [], frameon=False)  # Disable the legend if it is not needed
plt.title('Total Revenue for Price Allocation')
plt.ylabel('Total Revenue ($)')
plt.tight_layout()
plt.show()
```

## Availability-Based Allocation Strategy

The Availability-Based Allocation strategy aims to maximize hotel room utilization by allocating guests to hotels with the most available rooms first. 
The idea here is to ensure that the largest hotels (those with the most rooms) are filled first, which helps optimize room utilization across all properties.

```python
# Availability-Based Allocation Strategy
def availability_allocation(hotel_data, guest_data):
    hotels_copy = hotel_data.sort_values('rooms', ascending=False).copy()  # Sort hotels by room availability (descending)
    allocation = []

    for guest_id in guest_data.index:
        for hotel_id, hotel in hotels_copy.iterrows():
            # Allocate to the hotel with the most available rooms
            if hotel['rooms'] > 0:
                allocation.append({
                    'guest_id': guest_id,
                    'hotel_id': hotel_id,
                    'final_price': float(hotel['price'] * (1 - guest_data.loc[guest_id, 'discount']))
                })
                hotels_copy.at[hotel_id, 'rooms'] -= 1  # Decrease room count
                break

    return allocation

# Execute Availability Allocation Strategy
availability_results = availability_allocation(hotels.copy(), guests.copy())

# Convert results to DataFrame
availability_results_df = pd.DataFrame(availability_results)

# Analysis of Availability Allocation Results
total_revenue = availability_results_df['final_price'].sum()
total_guests_allocated = len(availability_results_df)
unique_hotels_used = availability_results_df['hotel_id'].nunique()

# Store results for further use
analysis_results['Availability Allocation'] = {
    'revenue': total_revenue, 
    'guests_allocated': total_guests_allocated,
    'unique_hotels_used': unique_hotels_used
}

# Visualization of Availability Allocation
plt.figure(figsize=(8, 6))

# Adding hue parameter to eliminate the deprecation warning
sns.barplot(x=['Availability Allocation'], y=[total_revenue], palette='cool', hue=['Availability Allocation'])
plt.legend([], [], frameon=False)  # Disable the legend if it is not needed
plt.title('Total Revenue for Availability Allocation')
plt.ylabel('Total Revenue ($)')
plt.tight_layout()
plt.show()
```

- **Availability-Based Allocation**: This strategy aimed to maximize room utilization by starting with hotels with the most available rooms. It helped efficiently utilize hotel capacity, especially for larger hotels, potentially leading to better balanced room usage.
